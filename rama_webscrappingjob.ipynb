```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from datetime import datetime
import time
import pandas as pd
import re
```

```python
path = "D:\Ramadhani\Data Analytics\chromedriver_win32\chromedriver.exe"
driver = webdriver.Chrome(path)
url = "https://www.linkedin.com/jobs/search?keywords=data%20analyst&location=indonesia&geoId=&trk=public_jobs_jobs-search-bar_search-submit&position=1&pageNum=0"

driver.maximize_window()
driver.get(url)

#Total Job that want to scrap
totaljobs= 450
# int(''.join(filter(str.isdigit, 
#                               driver.find_element(By.CSS_SELECTOR, 'h1>span').get_attribute('innerText')
#                               )
#                       )
#               )
n_scroll = int(totaljobs/25)+1
print(n_scroll)
i = 1
driver.execute_script("return document.body.scrollHeight")
while i <= n_scroll:
    driver.execute_script("return document.body.scrollHeight")
    time.sleep(4) # wait for 2 seconds
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);") #scroll to the bottom of page
    time.sleep(5)
    i = i + 1
    try:
        button = driver.find_element(By.XPATH,'//*[@id="main-content"]/section[2]/button') 
        #time.sleep(4)
        button.click()
        time.sleep(3)
        print("load more click")
    except:
        driver.execute_script("return document.body.scrollHeight")
        time.sleep(4)
```

```python
print ("total jobs :")
jobs = driver.find_element(By.CLASS_NAME,"jobs-search__results-list").find_elements(By.TAG_NAME,'li') # return a list
print(len(jobs))
```

```python
job_title= []
company_name = []
loc = []
date = []
joblink = []

i = 0
for job in jobs:
  job_title.append(job.find_element(By.CSS_SELECTOR,'h3').get_attribute('innerText'))
  
  company_name.append(job.find_element(By.CLASS_NAME,"base-search-card__subtitle").get_attribute('innerText'))
  
  loc.append(job.find_element(By.CLASS_NAME,"job-search-card__location").get_attribute('innerText'))

  date.append(job.find_element(By.TAG_NAME ,'time').get_attribute('datetime'))

  joblink.append(job.find_element(By.TAG_NAME,"a").get_attribute('href'))
```

```python
print('Length Job Title: ', len(job_title))
print('Length Company Name: ', len(company_name))
print('Length Location: ', len(loc))
print('Length Date Posted: ', len(date))
print('Length Company job link: ', len(joblink))
```

```python
jcc = '/html/body/div[1]/div/main/section[2]/ul/li[53]/a' #Job Click Path the url
jc = jobs[53].find_element(By.XPATH, jcc).click()
```

```python
jd = []
seniority = []
empy_type = []
jf = []
ind = []

for i in range(len(jobs)):
  print('Iteration Number: ', i)
  try: # Click the job to see the desc
    jcp = f"/html/body/div[1]/div/main/section[2]/ul/li[{i+1}]/div/a" #Job Click Path
    jc = job.find_element(By.XPATH, jcp).click() #Job Click
  except:
    jcp = f'/html/body/div[1]/div/main/section[2]/ul/li[{i+1}]/a' #Job Click Path
    jc = job.find_element(By.XPATH, jcc).click() #Job Click
  
  print('Job Click Path: ', jcp)
  jc = job.find_element(By.XPATH, jcp).click() #Job Click
  time.sleep(5)
  
  ##Return description of the job to the list
  try:
        jd_path = '/html/body/div[1]/div/section/div[2]/div/section[1]/div/div/section'
        jd0 = job.find_element(By.XPATH,jd_path).get_attribute('innerText')
  except:
        jd_path = '/html/body/div[1]/div/section/div[2]/div/section[2]/div/div/section/div'
  
  print('Job Desc Path: ',jd_path)
  jd0 = job.find_element(By.XPATH,jd_path).get_attribute('innerText')
  jd.append(jd0)

  jdp1 = '/html/body/div[1]/div/section/div[2]/div/section[1]/div/ul'
  ##Return seniority of the job to the list
  try:
    sp = jdp1 + '/li[1]/span' #Seniority Path
    seniority0 = job.find_element(By.XPATH,sp).get_attribute('innerText')
    print('Seniority XPath',sp)
    seniority.append(seniority0)
  except:
    seniority.append('')
  
  
  ##Return employment type of the job to the list
  try:
    etp = jdp1 + '/li[2]/span'#employment type path
    et0 = job.find_element(By.XPATH,etp).get_attribute('innerText')
    print('Employment Type Path: ', etp)
    empy_type.append(et0)
  except:
    empy_type.append('')


  ##Return job function of the job to the list 
  try:
    jfp = jdp1 + '/li[3]/span'#job function path
    jf0 = job.find_element(By.XPATH,jfp).get_attribute('innerText')
    print('Job Function Path: ', jfp)
    jf.append(jf0)
  except:
    jf.append('')


  ##Return Industries of the job to the list
  try:
    ip = jdp1 + '/li[4]/span'#Industries path
    ind0 = job.find_element(By.XPATH,ip).get_attribute('innerText')
    print('Industries Path: ', ip)
    ind.append(ind0)
  except:
    ind.append('')
  print('---------------------------------------------------------')
```

```python
print('Length Job Desc: ', len(jd))
print('Length Seniority: ', len(seniority))
print('Length employment type: ', len(empy_type))
print('Length Job Function: ', len(jf))
print('Length Industries: ', len(ind))
```

```python
js = pd.DataFrame({'Title': job_title,
                   'Company': company_name,
                   'Location': loc,
                   'Date': date,
                   'Description' : jd,
                   'Level': seniority,
                   'Type': empy_type,
                   'Function': jf,
                   'Industry': ind,
                   'Link': joblink
                  })
```
